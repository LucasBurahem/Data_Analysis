```python
import pandas as pd
from sklearn.preprocessing import normalize
```

```python
df = pd.read_json('data.json')
```

```python
itemList= 'ItemList'
productID = 'ProductID'
```

```python
df.head()
```

Expandir listas presentes em 'ItemList' para novas linhas

```python
df = df.explode(itemList)
```

```python
df.head(10)
```

Transformar cada chave no dicionario em "ItemList" em uma nova coluna

```python
df = pd.concat([df, df[itemList].apply(pd.Series)], axis=1).drop(columns=[itemList])
```

```python
df.reset_index(inplace=True)
df = df.drop('index', axis=1)
```

```python
df.head(10)
```

Foi criada uma nova colunas "ProductID" a partir do index para cada produto

```python
df[productID] = df.index
```

```python
df.head()
```

A dataframe foi divido em dois respeitando as 3 formas normais de banco de dados.

```python
nfe_data_columns = ['CreateDate', 'EmissionDate', 'Discount', 'NFeNumber', 'NFeID', 'ProductID', 'Quantity', 'Value']
nfe_products_columns = [productID, 'ProductName']
```

```python
df_nfe = df[df.columns[df.columns.isin(nfe_data_columns)]]
df_products = df[df.columns[df.columns.isin(nfe_products_columns)]]
```
